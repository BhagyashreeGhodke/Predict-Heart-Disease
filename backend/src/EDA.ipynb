# **Outliers Detection**

```python
import numpy as np

def detect_outliers_iqr(numerical_cols):
    outlier_counts = {}
    total_outliers = 0

    for col in numerical_cols:
        Q1 = balance_data[col].quantile(0.25)
        Q3 = balance_data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers
        outliers = balance_data[(balance_data[col] < lower_bound) | (balance_data[col] > upper_bound)]
        outlier_counts[col] = outliers.shape[0]
        total_outliers += outliers.shape[0]

    return outlier_counts, total_outliers
```

```python
numerical_cols = ['BMI', 'MentHlth', 'PhysHlth', 'Age']

outlier_counts, total_outliers = detect_outliers_iqr(numerical_cols)

print("Outliers per numerical column:")
for col, count in outlier_counts.items():
    print(f"{col}: {count}")

print(f"\nTotal outliers detected: {total_outliers}")
```

## **Countplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_countplots(df):
    sns.set(style="whitegrid")

    # 1. Heart Disease by High Blood Pressure
    plt.figure(figsize=(8, 5))
    sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df)
    plt.title('Heart Disease by High Blood Pressure Status')
    plt.show()

    # 2. Heart Disease by High Cholesterol
    plt.figure(figsize=(8, 5))
    sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df)
    plt.title('Heart Disease by High Cholesterol Status')
    plt.show()

    # 3. Heart Disease by Diabetes Status
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Diabetes', hue='HeartDiseaseorAttack', data=df)
    plt.title('Heart Disease by Diabetes Status')
    plt.show()

    # 4. Heart Disease by Income Level
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Income', hue='HeartDiseaseorAttack', data=df)
    plt.title('Heart Disease by Income Level')
    plt.xticks(rotation=45)
    plt.show()

    # 5. Heart Disease by Education Level
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Education', hue='HeartDiseaseorAttack', data=df)
    plt.title('Heart Disease by Education Level')
    plt.xticks(rotation=45)
    plt.show()

plot_countplots(df)
```

The countplots generated visually represent the distribution of categorical or ordinal features and how these distributions differ based on whether an individual has Heart Disease or Attack (`HeartDiseaseorAttack`).

Here's what each plot generally indicates:
- **Income vs Heart Disease**:
  Individuals with **lower income levels (1 to 5)** show a **higher count of heart disease cases (label 1)**.
  As income increases, the number of healthy individuals (label 0) increases significantly.

- **High Blood Pressure (HighBP) vs Heart Disease**:
  People with **high blood pressure (HighBP = 1)** are **less likely to be healthy**. Majority of them are diagnosed with heart disease.

- **High Cholesterol (HighChol) vs Heart Disease**:
  A large proportion of individuals with **high cholesterol levels (HighChol = 1)** are more likely to have heart disease.

- **Diabetes vs Heart Disease**:
  Those with diabetes status as **1 or 2** (i.e., prediabetic/diabetic) are significantly more likely to have heart disease compared to non-diabetics.

- **Education Level vs Heart Disease**:
  People with **higher education (level 5 or 6)** have a **lower occurrence** of heart disease compared to those with **lower education levels** (levels 1–3).

# **What the Boxplots Show**

The boxplots show the distribution of each numerical feature against the HeartDiseaseorAttack variable (which indicates whether a person has heart disease or not):

 - The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data.
 - The line inside the box is the median (the middle value).
 - The "whiskers" extend from the box to the minimum and maximum values within 1.5 times the IQR.
 - Any points beyond the whiskers are considered potential outliers.

- BMI vs. HeartDiseaseorAttack: If the boxplot for BMI shows that the median BMI is higher for individuals with Heart Disease or Attack compared to those without, it suggests a positive relationship between BMI and the likelihood of heart disease. You might also observe more outliers with higher BMI values in the heart disease group.
- GenHlth vs. HeartDiseaseorAttack: The boxplot for GenHlth (General Health) will likely show a trend where individuals with Heart Disease or Attack tend to report poorer general health (higher GenHlth values). The boxes and whiskers for the heart disease group might be shifted towards higher values compared to the no heart disease group.
- MentHlth vs. HeartDiseaseorAttack: This boxplot shows the distribution of the number of days of poor mental health. If individuals with heart disease have a higher median number of poor mental health days, or a wider spread of values towards more days, it suggests a potential link between mental health and heart disease.
- PhysHlth vs. HeartDiseaseorAttack: Similar to MentHlth, this boxplot for PhysHlth (Physical Health) can indicate if individuals with heart disease tend to report more days of poor physical health.
